# Introduction to big data

Workflow for working with big data
- big data -> better models -> higher precision

Process of using data streams
- integration -> personalization -> precision

Three types of data sources
- machine, people, organization
- people data is hard to process because velocity of data is fast and they tend to be very unstructured

Data silos
- data spread out between departments within an organization
- bad because they are unsynchronized and invisible data

Characteristics of big data
- volumn: sheer size of data
- variety: complexity. structural (ex. formats) / media (ex. how data is delivered) / semantic (ex. units) / availability (ex. real-time)
- velocity: speed at creating/storing/analyzing data
- veracity: quality
- valence: connectedness
- value: benefits to business

## Hadoop

HDFS (Hadoop Distributed File System)
- scalability / reliability
- spread files to differnt nodes for parallel access
- partitioning and replication for fault tolerance (node failure doesn't cause files to become unaccessible)
- NameNode for metadata (1 node per cluster), DataNode for block storage (each node per cluster)

YARN (Yet Another Resource Negotiator)
- resource manager for Hadoop
- while node manager is in charge of each machine, resource manager decides who gets what
- allows applications to run on the same Hadoop cluster

MapReduce
- programming model for Hadoop ecosystem
- relies on YARN
- not good for frequently changing data
- consists of mapping, shuffling, and reducing
- delegates large tasks to lots of nodes and aggregates the result

Weakness
- small datasets
- advanced algorithms
- infrastrutural replacement
- task level parallelism
- random data access

Hadoop commands
```
hadoop fs -copyFromLocal <input_text> 
hadoop fs -ls
hadoop jar /usr/jar/hadoop-examples.jar <command> <input_text> out
hadoop fs -copyToLocal out/part-r-00000 local.txt
hadoop -cp <file1> <file2>
hadoop -rm <file1>
<command> can be "wordcount", "wordmedian", etc
```
